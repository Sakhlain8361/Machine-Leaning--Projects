```python
data = pd.get_dummies(data, columns=['diagnosis'], prefix = 'diagnosis', drop_first=True)
```

```python
data
```

```python
data['diagnosis_M'] = data['diagnosis_M'].astype(int)
```

```python
data
```

```python
x = data.drop(['diagnosis_M'], axis = 1)
y = data['diagnosis_M']

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=60)
print("x_train",x_train.shape)
print("x_test",x_test.shape)
print("y_train",y_train.shape)
print("y_test",y_test.shape)
```

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors=15,p=2,metric='euclidean')

knn
```

```python
model = knn.fit(x_train, y_train)
model
```

```python
y_pred = model.predict(x_test)
y_pred
```

```python
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_test, y_pred)

print("The KNN confusion matrix :")
print(cm)
```

```python
from sklearn.metrics import accuracy_score

aa = accuracy_score(y_test, y_pred)*100

print("The KNN Accuracy Score:",aa)
```

```python
from sklearn.metrics import classification_report
crr = classification_report(y_test, y_pred,zero_division=0)
print("The classification report :")
print(crr)
```
